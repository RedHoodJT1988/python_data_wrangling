## Exercise 2.03: Implementing a Stack Using User-Defined Methods

```python
# create two new functions stack_push and stack_pop
def stack_push(s, value):
    return s + [value]
def stack_pop(s):
    tos = s[-1]
    del s[-1]
    return tos
# create an empty stack
url_stack = []
url_stack
```

```python
# create a string to analyze so that we can push the URLs in the stack one by one
wikipedia_datascience = """Data science is an interdisciplinary field that uses scientific methods, processes, algorithms and systems to extract knowledge [https://en.wikipedia.org/wiki/Knowledge] and insights from data [https://en.wikipedia.org/wiki/Data] in various forms, both structured and unstructured,similar to data mining [https://en.wikipedia.org/wiki/Data_mining]""" 
```

```python
# find the length of the string
len(wikipedia_datascience)
```

```python
wd_list = wikipedia_datascience.split()
wd_list
```

```python
# find the length of the list
len(wd_list)
```

```python
# loop through each word and check it is a URL
for word in wd_list:
    if word.startswith("[https://"):
        url_stack = stack_push(url_stack, word[1:-1])
        print(word[1:-1])
```

```python
# print the value
print(url_stack)
```

```python
# iterate over the list and print the URls
for i in range(0, len(url_stack)):
    print(stack_pop(url_stack))
```

```python
# print one more time to make sure the stack is empty
print(url_stack)
```

# Lambda Expressions

## Exercise 2.04: Implementing a Lambda Expression

```python
# import math package
import math

# define two functions my_sine and my_cosine
def my_sine():
    return lambda x: math.sin(math.radians(x))
def my_cosine():
    return lambda x: math.cos(math.radians(x))

# define sine and cosine
sine = my_sine()
cosine = my_cosine()
math.pow(sine(30), 2) + math.pow(cosine(30), 2)
```

## Exercise 2.07: Implementing a Queue in Python

```python
# create a Python queue with plain list methods
# record the time the append operation in the queue data structure takes
# we will use the %%time command
%time
queue = []
for i in range(0, 100000):
    queue.append(i)
print("Queue created")
queue
```

```python
for i in range(0, 100000):
    queue.pop(0)
print("Queue emptied")
```

```python
# implement the same queue with deque data structure
%time
from collections import deque
queue2 = deque()
for i in range(0, 100000):
    queue2.append(i)
print("Queue created")
for i in range(0, 100000):
    queue2.popleft()
print("Queue emptied")
```

```python
from itertools import permutations, dropwhile
permutations?
dropwhile?
```

```python
# use the permutations package to generate all possibile three-digit numbers using 1, 2, and 3
permutations(range(3))
```

```python
# loop over the iterator expression generated before,
# use the print method to print each element returned by the iterator
for number_tuple in permutations(range(3)):
    print(number_tuple)
    assert isinstance(number_tuple, tuple)
```

```python
# use dropwhile to remove the 0s from the tuple
for number_tuple in permutations(range(3)):
    print(list(dropwhile(lambda x: x <= 0, number_tuple)))
```

```python
# create a function that passes the list generated from dropwhile
import math
def convert_to_number(number_stack):
    final_number = 0
    for i in range(0, len(number_stack)):
        final_number += (number_stack.pop() * (math.pow(10, i)))
    return final_number
for number_tuple in permutations(range(3)):
    number_stack = list(dropwhile(lambda x: x <= 0, number_tuple))
    print(convert_to_number(number_stack))
```

## Activity 2.02: Designing Your Own CSV Parser

```python
# import zip_longest from itertools
from itertools import zip_longest
# define a function that contains `header`, `line`, and `fillvalue` as `None` and add it to a dictionary
def return_dict_from_csv(header, line):
    # Zip them
    zipped_line = zip_longest(header, line, fillvalue=None)
    # use dict comprehension to generate the final dict
    ret_dict = {kv[0]: kv[1] for kv in zipped_line}
    return ret_dict
# open the sales_record.csv file using read mode inside a with block
with open('./sales_record.csv', 'r') as fd:
    first_line = fd.readline()
    header = first_line.replace("\n", "").split(",")
    for i, line in enumerate(fd):
        line = line.replace("\n", "").split(",")
        d = return_dict_from_csv(header, line)
        print(d)
        if i > 10:
            break
```
